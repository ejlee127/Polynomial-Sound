```python
import simpleaudio as sa
import soundfile as sf
```

```python
with open("cyclo_15_43.txt", "r") as ft:
    coef = ft.readlines()
coefs = list(map(lambda st: int(st.split(',')[0]), coef))
print(len(coefs))

sr = 44100
note = np.array(coefs + coefs + coefs)
signal = note.astype(np.float)

# Display
librosa.display.waveplot(signal)

# Start playback
play_obj = sa.play_buffer(signal, 1, 2, sr)

# Wait for playback to finish before exiting
play_obj.wait_done()

# Write to a file
sf.write("cyclo_15_43.wav", signal, sr)
```
